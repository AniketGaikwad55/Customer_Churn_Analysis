### Replacing blank with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value==1:
        return "Yes"
    else:
        return "No"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

###converted 0 and 1 values of seniorecitizen to yes/no to make it easier to under stand

```python
plt.figure(figsize=(6,6))
ax=sns.countplot(x = 'Churn',data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(6,6))
gb=df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of churned Customers",fontsize=10)
plt.show()
```

#From the given pie chart we can conclude that 26.54% of our customers have churned out.

```python
plt.figure(figsize=(6,6))
sns.countplot(x = 'gender',data = df,hue='Churn')
plt.title("Churn by Gender",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(6,6))
ax=sns.countplot(x = 'SeniorCitizen',data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by SeniorCitizen",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x = 'SeniorCitizen',data = df,hue='Churn')
plt.title("Churn by SeniorCitizen",fontsize=10)
plt.show()
```

```python
churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
churn_percent = churn_counts.div(churn_counts.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(6, 6))

# Stack bars
bottom = 0
for churn_status in churn_percent.columns:
    ax.bar(churn_percent.index, churn_percent[churn_status], 
           label=churn_status, 
           bottom=bottom)
    bottom += churn_percent[churn_status]

# Add percentage labels
for i, (index, row) in enumerate(churn_percent.iterrows()):
    cum_height = 0
    for churn_status in churn_percent.columns:
        height = row[churn_status]
        ax.text(i, cum_height + height / 2, f'{height:.1f}%', 
                ha='center', va='center', fontsize=8)
        cum_height += height

# Customize plot
ax.set_xticks([0, 1])
ax.set_xticklabels(['Not Senior', 'Senior'])
ax.set_ylabel('Percentage')
ax.set_title('Churn by SeniorCitizen (Percentage)', fontsize=10)
ax.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
# List of columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Determine number of rows and columns for subplots
n_cols = 3
n_rows = -(-len(cols) // n_cols)  # Ceiling division

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
axes = axes.flatten()  # Flatten in case of single row/column

# Plot each countplot
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue='Churn', ax=axes[i])
    axes[i].set_title(f'Churn by {col}', fontsize=10)
    axes[i].tick_params(axis='x', rotation=45)

# Remove any unused axes
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#The majority of customers who do not churn tend to haveservices like Phone services,Internet Services(perticularly DSL), and online security enable for services like online backup,Techsupport,and streaming TV,churn rates are noticeably higher when these services are not use are unavailable.
